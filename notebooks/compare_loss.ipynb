```python
import matplotlib.pyplot as plt
import json
  
# Opening JSON file

lines_log1 = open('/home/BI/luunvt/image_retrieval/mmfewshot/work_dirs/pipeline_v1/fine_tune_2/3190/tfa_r101_fpn_voc-split1_5shot-fine-tuning_3190_5000iter/20230407_161626.log.json', 'r').readlines()
lines_log2 = open('/home/BI/luunvt/image_retrieval/mmfewshot/work_dirs/pipeline_v1/fine_tune_4/3190/tfa_r101_fpn_voc-split1_5shot-fine-tuning_3190_5000iter/20230410_092533.log.json', 'r').readlines()

loss_cls_log1 = [json.loads(line)['loss_cls'] for line in lines_log1[1:]]
loss_cls_log2 = [json.loads(line)['loss_cls'] for line in lines_log2[1:]]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,6))

ax1.plot(loss_cls_log1, 'b')
ax1.plot(loss_cls_log2, 'g')
ax1.set_title("loss_cls")
ax1.legend(['fine tune 1', 'fine tune 2'])

loss_bbox_log1 = [json.loads(line)['loss_bbox'] for line in lines_log1[1:]]
loss_bbox_log2 = [json.loads(line)['loss_bbox'] for line in lines_log2[1:]]

ax2.plot(loss_bbox_log1, 'b')
ax2.plot(loss_bbox_log2, 'g')
ax2.set_title("loss_bbox")
ax2.legend(['fine tune 1', 'fine tune 2'])

plt.show()
```
